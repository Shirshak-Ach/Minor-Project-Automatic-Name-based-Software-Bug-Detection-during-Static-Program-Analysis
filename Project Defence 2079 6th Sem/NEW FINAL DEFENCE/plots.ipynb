```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import math
  
# Using Numpy to create an array X
epoch = range(1, 9)

  
# Plotting both the curves simultaneously
plt.plot(epoch, train_loss, color='r', label='train loss')
plt.plot(epoch, val_loss, color='g', label='validation loss', linestyle='dashed')
plt.plot(epoch, test_loss, color='b', label='test loss', linestyle='dotted')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Epoch")
plt.ylabel("Loss")
# plt.title("Epoch vs Loss Function Swap")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()

plt.xlim(0, 10)
plt.ylim(0, 1)
  
# To load the display window
plt.show()
plt.savefig("Epoch vs Loss Function Swap.png")
```
