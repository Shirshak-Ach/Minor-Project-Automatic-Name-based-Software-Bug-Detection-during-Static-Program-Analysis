```python
swappable_operator = {
    '==': ['>', '<', '!=', '>=', '<=', '='],
    '<': ['>', '==', '!=', '>=', '<=', '<<', '^', '>>'],
    '+': ['-', '*', '/', '%', '+='],
    '*': ['+', '-', '/', '%', '*='],
    '-': ['+', '*', '/', '%', '-='],
    '!=': ['>', '<', '==', '>=', '<=', '='],
    '>': ['==', '<', '!=', '>=', '<=', '<<', '^', '>>'],
    '&': ['|', '^', '<<', '>>', '&&', '||'],
    '>=': ['>', '<', '!=', '==', '<=', '<<', '^', '>>'],
    '/': ['+', '-', '*', '%', '/='],
    '<=': ['>', '<', '!=', '>=', '==', '<<', '^', '>>'],
    '&&': ['||', '&', '|'],
    '<<': ['|', '&', '^', '>>', '&&', '||', '<=', '<'],
    '>>': ['|', '&', '^', '<<', '&&', '||', '>=', '>'],
    '||': ['&&', '|', '&', '/'],
    '%': ['+', '-', '*', '/', '<', '>', '>=', '<=', '!=', '%='],
    '|': ['&', '^', '<<', '>>', '&&', '||', '/'],
    '^': ['|', '&', '<<', '>>', '&&', '||', '<', '>']
}
```

```python
df = df.loc[(~df.operator.isin([",", "="]))]
df = df.reset_index(drop=True)
df
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df = df[df["operator"].isin(swappable_operator.keys())]
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.1, random_state=42)

df_train = df_train.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
def generate_random_swappable_operator(operator):
    possible_swaps = swappable_operator[operator]
    rd = random.randint(0, len(possible_swaps) - 1)
    return possible_swaps[rd]
```

```python
generate_random_swappable_operator("<=")
```

```python
print("Total:", len(df_train.index))

neg_train_array = []
count = 0

for index, row in df_train.iterrows():
    row["operator"] = generate_random_swappable_operator(row["operator"])
    row["labels"] = 1
    
    neg_train_array.append(row.values)
    
    count += 1
    if count % 25000 == 0:
        print(count)

df_train_neg = pd.DataFrame(neg_train_array, columns=df.columns)
df_train = pd.concat([df_train, df_train_neg])
df_train
```

```python
print("Total:", len(df_test.index))

neg_test_array = []
count = 0

for index, row in df_test.iterrows():
    row["operator"] = generate_random_swappable_operator(row["operator"])
    row["labels"] = 1
    
    neg_test_array.append(row.values)
    
    count += 1
    if count % 25000 == 0:
        print(count)

df_test_neg = pd.DataFrame(neg_test_array, columns=df.columns)
df_test = pd.concat([df_test, df_test_neg])
df_test
```
