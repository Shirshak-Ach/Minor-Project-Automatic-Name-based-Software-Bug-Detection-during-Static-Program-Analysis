```python
df['file'] = df['file'].apply(lambda x: "D:\\"+x.replace("/","\\"))
```

```python
tp = []
tp_fn = []
fp = []
fp_fn = []
fn = []
fn_fn = []
tn = []
tn_fn = []
```

```python
df.index
```

```python
total = 5

for i in df.index:
    try:
        if df['labels'][i]==0 and df['prediction'][i] == 0: #TN
            with open(df['file'][i], "r") as f:
                content = str(f.read())
                if content.count("\n") <= 50:
                    tn.append(content)
                    tn_fn.append(df['function_name'][i] + " " + df['arg1'][i] + " " + df['arg2'][i])
        elif df['labels'][i]==0 and df['prediction'][i] == 1: #FP
            with open(df['file'][i], "r") as f:
                content = str(f.read())
                if content.count("\n") <= 50:
                    fp.append(content)
                    fp_fn.append(df['function_name'][i] + " " + df['arg1'][i] + " " + df['arg2'][i])
        elif df['labels'][i]==1 and df['prediction'][i] == 0: #FN
            with open(df['file'][i], "r") as f:
                content = str(f.read())
                if content.count("\n") <= 50:
                    fn.append(content)
                    fn_fn.append(df['function_name'][i] + " " + df['arg1'][i] + " " + df['arg2'][i])
        else:
            with open(df['file'][i], "r") as f: #TP
                content = str(f.read())
                if content.count("\n") <= 50:
                    tp.append(content)
                    tp_fn.append(df['function_name'][i] + " " + df['arg1'][i] + " " + df['arg2'][i])

        if(len(tp) >= total and len(fp) >= total and len(tn) >= total and len(fn) >= total):
            break

        print(len(tp), len(fp), len(tn), len(fn))
    
    except:
        pass
```

```python
aaaa = 5
```

```python
for i in range(min(len(tp), aaaa)):
    with open("tp" + str(i) + ".c", 'w') as f:
        f.write(str(tp_fn[i]))
        f.write("\n----------------------------------------------\n")
        f.write(str(tp[i]))

for i in range(min(len(fn), aaaa)):
    with open("fn" + str(i) + ".c", 'w') as f:
        f.write(str(fn_fn[i]))
        f.write("\n----------------------------------------------\n")
        f.write(str(fn[i]))

for i in range(min(len(tn), aaaa)):
    with open("tn" + str(i) + ".c", 'w') as f:
        f.write(str(tn_fn[i]))
        f.write("\n----------------------------------------------\n")
        f.write(str(tn[i]))
        
for i in range(min(len(fp), aaaa)):
    with open("fp" + str(i) + ".c", 'w') as f:
        f.write(str(fp_fn))
        f.write("\n----------------------------------------------\n")
        f.write(str(fp[i]))
```
