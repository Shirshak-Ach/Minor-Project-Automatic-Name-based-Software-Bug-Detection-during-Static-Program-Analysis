```python
def get_c_headers_count(root_dir, header_count):
    count = 0
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".c"):
                count += 1
                file_path = os.path.join(root, file)
                with open(file_path, 'rb') as f:
                    content = str(f.read())
                    for element in re.findall('#include\s+[<|"]([^>"]*)[>|"]', content):
                        if element in header_count.keys():
                            header_count[element] += 1
                        else:
                            header_count[element] = 1
    print(count)
```

## Checking compilation

```python
import subprocess

result = subprocess.run(["clang", "-o", "program", "main.c"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

if result.returncode == 0:
    print("Program compiled successfully.")
else:
    print("Compilation failed with the following error:\n", result.stderr.decode())
```

```python
def get_compilation_count(root_dir):
    compile_count, error_count, total = 0, 0, 0
    
    with open("correct_code_list.txt", 'a') as correct_code_list:
        correct_code_list.write("path\n")
        
        for root, dirs, files in os.walk(root_dir):
            for file in files:
                if file.endswith(".c"):
                    total += 1
                    file_path = os.path.join(root, file)
                    with open(file_path, 'rb') as f:
                        program = str(f.read())

                        result = subprocess.run(["clang", "-o", "program", f.name], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

                        if result.returncode == 0:
                            compile_count += 1
                            correct_code_list.write(f.name + "\n")
                        else:
                            error_count += 1
            
                if total %  10000 == 0:
                    print("Total:", total, "||", "Compile count:", compile_count, "||", "Error count:", error_count)
    return {"compile_count": compile_count, "error_count": error_count, "total": total}
```
